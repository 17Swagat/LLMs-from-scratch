## **Simple Tokenizer (V1)**

```python
with open(TEXTFILE_PATH, 'r') as file:
    raw_text = file.read()
raw_text
```

```python
#  text -> tokens -> token ids
preprocessed = re.split(r'([,.:;?_!"()\']|--|\s)', raw_text) 
vocab = [token for token in sorted(set(preprocessed)) if token.strip()]
print(len(vocab))
print(vocab)
```

```python
# NOTE: The following tokenizer is not meant for handling spaces, it strips out the space present in the input text for the encoder.
# NOTE: Its not the same one as in the book, but behaves in the same way.

class SimpleTokenizer:
    def __init__(self, vocab):
        '''vocab: "List of unique strings present in the training raw_text" '''
        self.vocab = vocab
        self.token_to_id = {token: idx for idx,token  in enumerate(vocab)}
        self.id_to_token = {idx: token for idx,token  in enumerate(vocab)}
    
    def encode(self, text):
        '''text -> token -> tokenId'''
        preprocessed = re.split(r'([,.?_!"()\']|\s)', text)
        tokenIDs = [self.token_to_id[token] for token in preprocessed if token.strip()]
        return tokenIDs
    
    def decode(self, tokenIds):
        '''tokenId -> token'''
        result = ' '.join([self.id_to_token[id] for id in tokenIds])
        return result
```

```python
simpleTokenizer1 = SimpleTokenizer(vocab)
tokenIds = simpleTokenizer1.encode('how are you doing?')
print(tokenIds)
print(simpleTokenizer1.decode(tokenIds))
```

```python
sample_ids = simpleTokenizer1.encode('And Among ! Among among this is ')
print(sample_ids)
print(simpleTokenizer1.decode(sample_ids))
```

```python
sample_ids = simpleTokenizer1.encode('and be ! Among among this is ')
print(sample_ids)
print(simpleTokenizer1.decode(sample_ids))
```

```python
text = """"It's the last he painted, you know," Mrs. Gisburn said with pardonable pride."""
ids = simpleTokenizer1.encode(text)
print(ids)
texts = simpleTokenizer1.decode(ids)
print(texts)
```

## **Simple Tokenizer (V1) => `From Book`**

```python
class SimpleTokenizerV1_FromBook:
    def __init__(self, vocab):
        self.str_to_int = vocab
        self.int_to_str = {i:s for s,i in vocab.items()}
    
    def encode(self, text):
        preprocessed = re.split(r'([,.:;?_!"()\']|--|\s)', text)
                                
        preprocessed = [
            item.strip() for item in preprocessed if item.strip()
        ]
        ids = [self.str_to_int[s] for s in preprocessed]
        return ids
        
    def decode(self, ids):
        text = " ".join([self.int_to_str[i] for i in ids])
        # Replace spaces before the specified punctuations
        text = re.sub(r'\s+([,.?!"()\'])', r'\1', text)
        return text
```

```python
vocab = sorted(set(
    re.split(r'([,.:;?_!"()\']|--|\s)', raw_text) # "From Book"
))
vocab = [text for text in vocab if text.strip()]
print(vocab)
vocab = {text: idx for idx, text in enumerate(vocab)} # { textToken : id }
print(vocab)
```

```python
simpleTokenizer1_FromBook = SimpleTokenizerV1_FromBook(vocab)
# Text from book
text = """"It's the last he painted, you know," Mrs. Gisburn said with pardonable pride."""
ids = simpleTokenizer1_FromBook.encode(text)
```

```python
class SimpleTokenizerV2:
    '''This tokenizer is meant to handle <Unknown> words.'''

    def __init__(self, vocabList):
        '''vocab: list of unique vocabs in the raw_text Training File.'''
        self.token2Id = {tokenText:tokenId for tokenId, tokenText in enumerate(vocabList)}
        self.Id2token = {tokenId:tokenText for tokenId, tokenText in enumerate(vocabList)}
    
    def encode(self, text):
        preprocessed = re.split(r'([,.:;?_!"()\']|--|\s)', text)
        preprocessed = [textToken for textToken in preprocessed if textToken.strip()]

        # Handling unknown Tokens
        preprocessed = [textToken if textToken in vocab else '<|unk|>' for textToken in preprocessed]
        preprocessed = [self.token2Id[token] for token in preprocessed]

        # all_tokens.extend(['<|endoftext|>', '<|unk|>'])
        # preprocessed =  [self.token2Id[textToken] for textToken in preprocessed if textToken.strip()]
        return preprocessed
    
    # MyVersion:
    def decode(self, ids):
        return ' '.join([self.Id2token[tokenId] for tokenId in ids])

    # BookVersion:
    # def decode(self, ids):
    #     text = " ".join([self.Id2token[i] for i in ids])
    #     # Replace spaces before the specified punctuations
    #     text = re.sub(r'\s+([,.:;?!"()\'])', r'\1', text)
    #     return text

simpleTokenizer2 = SimpleTokenizerV2(vocab)
text = """"It's the last he painted, you know," 
        Mrs. Gisburn said with pardonable pride. Hello Hello ZFSFS"""

# NOTE: "You can see This tokenizer is able to Handle Unknown Words"
TOKEN_IDS = simpleTokenizer2.encode(text)
print(TOKEN_IDS)
```
