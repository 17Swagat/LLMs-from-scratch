```python
import torch
```

```python
# input-embedding:
inputs = torch.tensor(
  [[0.43, 0.15, 0.89], # Your     (x^1)
   [0.55, 0.87, 0.66], # journey  (x^2)
   [0.57, 0.85, 0.64], # starts   (x^3)
   [0.22, 0.58, 0.33], # with     (x^4)
   [0.77, 0.25, 0.10], # one      (x^5)
   [0.05, 0.80, 0.55]] # step     (x^6)
)
```

```python
# GOAL: To have the context-vector z_2: 
# "z_2 is a rich representation of x_2(same shape) with information about other tokens incorporated into it."

# Find `attention_score_2`:
query = inputs[1]
attention_score_2 = torch.empty(inputs.shape[0])
for i, x_i in enumerate(inputs):
    attention_score_2[i] = torch.dot(query, x_i)

print('Attention-Score BEFORE Softmax:\n', attention_score_2)

# Normalize `attention_score_2` with softmax:
attention_score_2 = torch.softmax(attention_score_2, dim=0) # becomes "attention-weights"

print('\nAttention-Score AFTER Softmax:\n', attention_score_2)

# Finding z_2
context_z2 = torch.zeros(query.shape)
for i, x_i in enumerate(inputs):
    context_z2 += x_i * attention_score_2[i]

context_z2 
```

```python
# input-embedding:
x = torch.tensor(
  [[0.43, 0.15, 0.89], # Your     (x^1)
   [0.55, 0.87, 0.66], # journey  (x^2)
   [0.57, 0.85, 0.64], # starts   (x^3)
   [0.22, 0.58, 0.33], # with     (x^4)
   [0.77, 0.25, 0.10], # one      (x^5)
   [0.05, 0.80, 0.55]] # step     (x^6)
)

x_2 = x[1]
attn_score_2 = x * x_2
attn_score_2 = torch.sum(attn_score_2, dim=-1)
attn_score_2
```

```python
attn_weight_2 = attn_score_2.softmax(dim=-1)
attn_weight_2
```

```python
# torch.zeros()
context_vector2 = torch.zeros(x_2.shape)
for i, x_i in enumerate(x):
    context_vector2 += (attn_weight_2[i] * x_i)
    # print(attn_weight_2[i] * x_i)
context_vector2
```

```python
context_z2_v2 = torch.zeros(x_2.shape)
for i, x_i in enumerate(x):
    context_z2_v2 += x_i * attn_weight_2[i]
```
